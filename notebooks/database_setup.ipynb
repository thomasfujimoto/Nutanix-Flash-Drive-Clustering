# Uploads the content of ssd_devices_info to sql datbase:

```python
import psycopg2

# Connection parameters
conn = psycopg2.connect(database="nutanix", 
                        user="postgres", 
                        host='172.25.221.34',
                        password="Senna",
                        port=1433)

try:
    # Create a cursor
    cur = conn.cursor()

    # Path to your CSV file
    csv_file_path = 'ssd_devices_info_20241016.csv'

    # Drop the table if it exists
    drop_table_query = "DROP TABLE IF EXISTS storage_devices;"
    drop_table_query = "DROP TABLE IF EXISTS ssd_devices_info;"
    cur.execute(drop_table_query)  # Drop the existing table
    conn.commit()  # Commit the drop operation

    # Create the table
    create_table_query = """
    CREATE TABLE ssd_devices_info (
        concord_id UUID,  -- No PRIMARY KEY constraint
        type VARCHAR(50),
        family VARCHAR(50),
        vendor VARCHAR(50),
        model VARCHAR(50),
        firmware VARCHAR(50),
        capacity_GiB FLOAT,  
        operating_pci_speed_GTs FLOAT,  
        operating_pci_width FLOAT,
        linkrate_Gbs FLOAT 
    );
    """
    
    # Execute the CREATE TABLE statement
    try:
        cur.execute(create_table_query)
        conn.commit()  # Commit the creation of the table
        print("Table created successfully.")
    except Exception as e:
        print(f"Error creating table: {e}")
        conn.rollback()  # Rollback if table creation fails

    # Load data from the CSV file
    with open(csv_file_path, 'r') as f:
        # Skip the header row (if your CSV has one)
        next(f)  
        # Copy the data into the table
        try:
            cur.copy_expert("COPY ssd_devices_info FROM STDIN WITH CSV", f)
            conn.commit()  # Commit the data load
            print("Data loaded successfully.")
        except Exception as e:
            print(f"Error loading data: {e}")
            conn.rollback()  # Rollback if data loading fails

except Exception as e:
    print(f"An unexpected error occurred: {e}")

finally:
    # Close the cursor and connection
    if cur:
        cur.close()
    if conn:
        conn.close()
```
